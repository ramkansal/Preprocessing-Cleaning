# **Handle Outliers**


1.   Using Z-Score



```python
from scipy import stats
# Select only numeric columns
numeric_df = df.select_dtypes(include=np.number)

# Calculate z-scores on numeric columns
df_zscores = np.abs(stats.zscore(numeric_df))

# Filter rows based on z-scores
# Keep rows where all z-scores are less than 3
filtered_df = df[(df_zscores < 3).all(axis=1)]
```
